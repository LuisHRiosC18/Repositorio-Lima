## Subset Sum Problem

Este problema consiste en encontrar dos subconjuntos de un conjunto de números enteros dado, tal que la suma de los elementos en ambos subconjuntos sea igual.

```python
import random
from itertools import combinations
```

```python
def fun(cantidad, rango):
    '''
    Esta función genera un conjunto de $cantidad$ números enteros en un
    rango de -k a k con k=rango que introduzcamos.
    '''
    lista= [random.randint(-rango,rango) for i in range(cantidad)]
    return set(lista)
```

```python
st = fun(20,10)
st
```

```python
def checa_cero(conjunto):
    k= list(conjunto)
    n= len(k)
    c= set(conjunto)
    for i in range(1,n+1):
        for sub in combinations(c, i):
            if sum(list(sub)) == 0:
                return set(sub)
    return False
```

```python
checa_cero(st)
```

```python
def checa_suma(conjunto):
    k= list(conjunto)
    n= len(k)
    c= set(conjunto)
    lid=[]
    for i in range(1,n+1):
        s=0
        for sub in combinations(c, i):
            lid.append(list(sub))
    for i in range(0, len(lid)+1):
        for j in range(i+1, len(lid)+1):
            if sum(lid[i]) == sum(lid[j]):
                return (set(lid[i]), set(lid[j]))
    return False
```

```python
st = fun(20,10)
checa_suma(st)
```
